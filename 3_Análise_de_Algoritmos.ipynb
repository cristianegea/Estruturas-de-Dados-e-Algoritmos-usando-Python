# 1. Análise de complexidade

## 1.1 Notação Big-O

Considerando que a função $T(n)$ representa o nº aproximado de passos requeridos pelo algoritmo para um input de tamanho $n$ e que haja uma função $f(n)$, definida para inteiros $n \geq 0$, tal que para alguma constante $c$ e alguma constante $m$,

$$
T(n) \leq cf(n)
$$

para todos os valores grandes do suficiente de $n \geq m$. Portanto, tal algoritmo é dito ter uma complexidade de tempo de $f(n)$ relativa ao número de operações requeridas. Em outras palavras, há um inteiro positivo $m$ e uma constante $c$ (constante de proporcionalidade) tal que, para todo $n \geq m$, $T(n) \leq cf(n)$.

A função $f(n)$ indica a taxa de crescimento do tempo de execução de um algoritmo à medida que o tamanho do input $n$ cresce. Para especificar a complexidade de tempo de um algoritmo, que executa na ordem de $f(n)$, utiliza-se a notação

$$
O(f(n))
$$

O objetivo é encontrar uma função $f(.)$ que fornece o limite superior mais apertado (mais baixo) ou limite para o tempo de execução do algoritmo. A notação big-O destina-se a indicar uma eficiência do algoritmo para grandes valores de $n$.

**Constante de proporcionalidade**

A constante de proporcionalidade é somente crucial quando 2 algoritmos possuem o mesmo $f(n)$. Geralmente, não há diferença quando comprados algoritmos cujas taxas de crescimento são de magnitudes diferentes.

**Construindo $T(n)$**

O número total de operações requeridas por um algoritmo pode ser calculada como a soma dos tempos requeridso para executar cada etapa.

$$
T(n) = f_1(n) + f_2(n) + \dots + f_k(n)
$$

As etapas que requerem tempo constante são geralmente omitidas visto que eles eventualmente tornam-se parte da constante de proporcionalidade.

**Escolhendo a função**

A função $f(n)$ utilizada para categorizar um algoritmo particular é escolhida para ser o termo determinante de $T(n)$.

**Classes de algoritmos**

Os algoritmos podem ser classificados com base em sua função big-O. As diversas classes são comumentemente chamadas com base no seu termo dominante.

* Algoritmo logarítmico: algoritmo com complexidade de tempo $O(\log_a n)$

* Algoritmo polinomial: algorito com complexidade $O(n^m)$
